```python
import turtle
import math

# Function to draw the polar curve
def draw_polar_curve():
    turtle.speed(0)
    turtle.color("blue")

    # Iterate over theta from 0 to 2*pi
    for theta in range(0, 628, 1):
        theta = theta / 100
        r = math.sin(3 * theta)
        x = r * math.cos(theta)
        y = r * math.sin(theta)
        turtle.goto(x * 100, y * 100)

    turtle.hideturtle()

# Set up the turtle screen
turtle.setup(width=800, height=800)
turtle.bgcolor("white")
turtle.title("Polar Curve: r = sin(3*theta)")

# Draw the polar curve
draw_polar_curve()

# Keep the window open
turtle.done()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter range
t_values = np.linspace(2, 4, 100)

# Calculate x and y values using the parametric equations
x_values = t_values ** 2
y_values = t_values - 2

# Plot the parametric curve
plt.plot(x_values, y_values, label='Parametric Curve')
plt.title('Parametric Curve: x = t^2, y = t - 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define t values
t_values = np.linspace(-10, 10, 400)

# Calculate x and y values using the parametric equations
x_values = t_values + 3
y_values = t_values / (t_values + 3)

# Plot the curve
plt.plot(x_values, y_values, label='Parametric Curve', color='blue')
plt.title('Curve Represented by Parametric Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Add arrows to indicate the direction of the curve
for i in range(len(t_values) - 1):
    dx = x_values[i + 1] - x_values[i]
    dy = y_values[i + 1] - y_values[i]
    plt.arrow(x_values[i], y_values[i], dx, dy, head_width=0.1, head_length=0.1, color='red')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ellipse parameters
a = 5  # semi-major axis
b = 4  # semi-minor axis

# Define the range of theta values
theta = np.linspace(0, 2*np.pi, 1000)

# Parametric equations for the ellipse
x = a * np.cos(theta)
y = b * np.sin(theta)

# Plot the ellipse
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Ellipse')

# Plot the vertices
plt.plot(0, 5, 'ro')  # Vertex at (0, 5)
plt.plot(0, -5, 'ro')  # Vertex at (0, -5)

# Plot the foci
plt.plot(0, 3, 'bo')  # Focus at (0, 3)
plt.plot(0, -3, 'bo')  # Focus at (0, -3)

# Label the vertices and foci
plt.text(0.2, 5.2, 'Vertex (0, 5)', fontsize=10, color='red')
plt.text(0.2, -5.5, 'Vertex (0, -5)', fontsize=10, color='red')
plt.text(0.2, 3.2, 'Focus (0, 3)', fontsize=10, color='blue')
plt.text(0.2, -3.5, 'Focus (0, -3)', fontsize=10, color='blue')

# Set plot limits and aspect ratio
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ellipse with Vertices and Foci')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
v0 = 1000  # initial velocity in ft/s
theta = 5  # launch angle in degrees
g = 32     # acceleration due to gravity in ft/s^2

# Convert angle to radians
theta_rad = np.radians(theta)

# Calculate time of flight
T = (2 * v0 * np.sin(theta_rad)) / g

# Calculate maximum height
H = ((v0 * np.sin(theta_rad)) ** 2) / (2 * g)

# Calculate horizontal distance
D = v0 * np.cos(theta_rad) * T

# Time array for plotting
t_values = np.linspace(0, T, num=100)

# Calculate x and y coordinates for the rocket's path
x_values = v0 * np.cos(theta_rad) * t_values
y_values = v0 * np.sin(theta_rad) * t_values - 0.5 * g * t_values ** 2

# Plot the rocket's path
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Rocket path')
plt.title('Rocket\'s Path')
plt.xlabel('Horizontal Distance (ft)')
plt.ylabel('Vertical Distance (ft)')
plt.axhline(0, color='black', linewidth=0.5)  # Ground level
plt.axvline(0, color='black', linewidth=0.5)  # Vertical line
plt.grid(True)
plt.legend()
plt.show()

# Print the calculated values
print("Time of flight (s):", round(T, 3))
print("Maximum height (ft):", round(H, 1))
print("Horizontal distance traveled (ft):", round(D))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations
def x(t):
    return 21 * t

def y(t):
    return -16 * t**2 + 36.37 * t + 5

# Generate time values
t_values = np.linspace(0, 2.4, 100)  # 100 time points from 0 to 2.4 seconds

# Calculate x and y values for each time point
x_values = x(t_values)
y_values = y(t_values)

# Plot the trajectory
plt.plot(x_values, y_values, label='Trajectory')
plt.title('Trajectory of the Baseball')
plt.xlabel('Horizontal Distance (ft)')
plt.ylabel('Vertical Distance (ft)')
plt.axhline(0, color='black', linewidth=0.5)  # Horizontal line at y=0 (ground level)
plt.legend()
plt.grid(True)
plt.show()
```
